# Proyecto Data Sciencs
El presente trabajo se basa en el análisis de un datasets proveniente del Hospital General de Massachusetts, que abarca el período comprendido entre 2011 y 2022. la base de datos incluye información demográfica de los pacientes, datos sobre su cobertura de salud y un registro detallado de las consultas médicas y los procedimientos que tuvieron. Esta información resulta de gran importancia para comprender los patrones de atención sanitaria, las características de la población asisitida y la distribución de los procedimientos médicos a lo largo del tiempo.


# Objetivo
El principal objetivo de este estudio es vincular la información demográfica de los pacientes con los procedimientos médicos que recibieron, con el fin de identificar relaciones significativas entre las características de los pacientes (género) y los tipos de procedimientos recibidos. Pudiendo así, identificar tendencias en la utilización de servicios, detectar desigualdades en el acceso a ciertos procedimientos y contar con evidencia suficiente para la planificación sanitaria y la gestión hospitalaria.
